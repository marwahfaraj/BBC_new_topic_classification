```python
# Import necessary libraries
import pandas as pd
from pycaret.classification import setup, compare_models, blend_models, finalize_model, predict_model, plot_model

# Load the data
data = pd.read_csv('data/bbc_news_text_complexity_summarization.csv')

# Select only the 'text' and 'labels' columns
data = data[['text', 'labels']]

# Display the first few rows of the dataset to check structure
print(data.head())

# Set up PyCaret for a classification task
# Assuming 'labels' is the target column containing the categories for classification
clf_setup = setup(data=data, 
                  target='labels',  # Target column for classification
                  session_id=123, 
                  use_gpu=True)  # Use GPU if available for faster processing

# Compare baseline models to select top performing ones
best_models = compare_models(n_select=3)

# Blend the top models
blended_model = blend_models(estimator_list=best_models)

# Finalize the model to make it ready for deployment
final_model = finalize_model(blended_model)

# Predict on the dataset to see model performance
predictions = predict_model(final_model)

# Display predictions
print(predictions.head())

# Plot the Confusion Matrix
plot_model(final_model, plot='confusion_matrix')

# Plot the ROC Curve
plot_model(final_model, plot='auc')

# Plot Feature Importance (if applicable)
plot_model(final_model, plot='feature')
```
